## Importing Dependencies

```
import pandas as pd
import tensorflow as tf
```

```
data = pd.read_csv("/content/drive/My Drive/Final Models/Datasets/imdb_r.csv")
```

```
data.head()
```

## Term frequency‚Äìinverse document frequency

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfv = TfidfVectorizer(stop_words='english')

tfidfv_matrix = tfidfv.fit_transform(data['Overview'])

tfidfv_matrix.shape
```

## Computing cosine similarity.

```
from sklearn.metrics.pairwise import linear_kernel

cosine_simi = linear_kernel(tfidfv_matrix, tfidfv_matrix)
```

```
# Reverse Map For Movies
indices = pd.Series(data.index, index=data['MovieName'])
```

```
# Fucntion To get recommendations
def get_recommendations(title, cosine_simi = cosine_simi):
    idx = indices[title]

    simi_scores = list(enumerate(cosine_simi[idx]))

    simi_scores = sorted(simi_scores, key=lambda x: x[1], reverse=True)

    simi_scores = simi_scores[1:11]

    movie_indices = [i[0] for i in simi_scores]

    return data['MovieName'].iloc[movie_indices]
```

## Final Recommendation

```
get_recommendations('Bigil')
```
