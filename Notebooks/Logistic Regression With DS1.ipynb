## Imports

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```
df = pd.read_csv('/content/drive/My Drive/Final Models/Datasets/qupic_dataset.csv')
```

```
df.head()
```

## Data Cleaning 

```
df.drop('Comedian',axis=1,inplace=True)
```

```
df.drop('SNo',axis=1,inplace=True)
```

```
df.drop('ProductionHouse',axis=1,inplace=True)
```

```
df.drop('Actress',axis=1,inplace=True)
```

```
df.drop('MovieName',axis=1,inplace=True)
```

```
df.head()
```

## One Hot Encoding

```
temp = pd.get_dummies(df['Director'],drop_first=True)
```

```
temp
```

```
temp2 = pd.get_dummies(df['ReleaseMonth'],drop_first=True)
```

```
temp2
```

```
temp3 = pd.get_dummies(df['Genre'],drop_first=True)
```

```
temp3
```

```
temp4 = pd.get_dummies(df['Actor'],drop_first=True)
```

```
temp4
```

```
temp5 = pd.get_dummies(df['Censorship'],drop_first=True)
```

```
temp5
```

```
temp6 = pd.get_dummies(df['TargetAudience'],drop_first=True)
```

```
temp6
```

```
data = pd.concat([temp,temp2,temp3,temp4,temp5,temp6,df['ViewerRating']],axis=1)
```

```
y = pd.get_dummies(df['HitOrFlop'],drop_first=True)
```

## Train Test Split

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(data, y, test_size=0.1, random_state=42)
```

## Logistic Regression

```
from sklearn.linear_model import LogisticRegression
```

```
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```
predictions = logmodel.predict(X_test)
```

## Model Evaluation

```
from sklearn.metrics import classification_report,confusion_matrix
```

```
print(classification_report(y_test,predictions))
```

```
print(confusion_matrix(y_test,predictions))
```
